## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
# using bar chart to plot earthquake magnitude for the varios geo locations
# draw a bar plot
df1 = pd.DataFrame(df, columns=["location", "magnitude"])
df1.plot(x = 'location', y = ['magnitude'] , kind ="bar", figsize=(26, 18));
plt.show()
```

```python
# using scatter plot to show earthquake magnitudes for various geo locations...
# draw a scatter plot
ax1 = plt.scatter(x=df['location'].astype(str), y=df['magnitude'], c ='DarkBlue', s= 10)
plt.xlabel('location')
plt.ylabel('magnitude')
```

```python
#checking the frequency of the earthquakes based on their magnitude.
df['magnitude'].value_counts().plot(kind='bar')
plt.xlabel('magnitude')
plt.ylabel('frequency')
```
